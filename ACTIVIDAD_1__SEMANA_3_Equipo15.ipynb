El almacenamiento de datos es aquel espacio en la nube que contiene todos los datos recolectados de diferentes fuentes ya sean de una manera estructurada o no.
Estas fuentes pueden ser de archivos de excel, csv, SQL, JSTOR, entre otras.
Sin duda uno de los retos más importantes para los científicos de datos es el preprocesamiento de los datos y la transformación de éstos ya que de esto depende el éxito del análisis.
Como bien se dice en este ámbito, si ingresas basura, saldrá basura y es por eso que la inversión de tiempo más grande es en el preprocesamiento de datos,y el entrenamiento de éstos para posteriormente modelarlos.
EL método para que el almacenamiento de datos funcione satisfactoriamente es el poder normalizar los datos ya que como éstos vienen de diferentes fuentes es importante que toda fuente sea estandarizada.
Cuando hablamos de una fuente estructurada, hablamos de datos que están por lo general especificados en una tabla formada por filas llaamdas también records que son las que contienen la información ya sea numérica o categórica asi como las columnas también llamadas campos o atributos que justamente son las clasficiaciones en las que la información está estructurada.
Por lo general, para unificar dos tablas se debe hacer a través de su columna principal ("primary key") la cual en cada record contiene un valor único que és el encargado de unir la información de una con la otra. Este "primary key" puede ser extraída también de una fuente alterna la cual la conocemos como "foreign key" y en conjunto con la columna principal pueden unir la información de acuerdo a como se solicita.
Por último, existen dos procesos para almacenar los datos los cuales se conocen por sus siglas ETL (Extract, Transform, Load y ELT (Extract, Load, Transform), ésta última ha sustituido a la primera en los últimos años ya que es un proceso que genera mayor practicidad y escalabilidad al momento de transformar la información. Es por eso que cuando se trabaja con diferentes fuentes es importante identificar si estamos hablando de una fuente primaria o una fuente que fue transformada anteriormente para poder entender el porque de los datos.
